```python
import yfinance as yf
import datetime as dt
import pandas as pd

# Hardcoding PRIA Inputs for testing
ticker = 'AAPL'                                   # Given by PRIA
ID_date = dt.date(2023,6,7)                       # Calculated using dt.datetime(now), when FRED identifes it as new stock from PRIA
split_date = dt.date(2023,6,15)                   # Given by PRIA **NOTE: this is the day that the stock trades on an adjusted basis**
split_waittime = (split_date - ID_date).days      # Number of days stock was identified before its split date 
end_date = split_date + dt.timedelta(weeks=3)     # Time at which we stop recording stock and don't care annymore **NOTE: using timedelta of 3 weeks on average gives about 2 weeks after split bc weekends lol**
split_mag = 10                                    # Given by PRIA
Rounding = True                                   # Given by PRIA

# Obtaining Yahoo finance historical data on given stock 
ticker_obj = yf.Ticker(ticker)
historical_data = ticker_obj.history(start=ID_date, end=(end_date), interval='1d')



                                                                 ##### TECHNICAL DATA #####
# Identifying Split Related Information                       
PreS_Close = ((ticker_obj.history(start=(split_date - dt.timedelta(days=1)), end=(split_date), interval='1d'))["Close"]).item() # Closing Price the evening before the split
PoS_Open = ((ticker_obj.history(start=(split_date), end=(split_date + dt.timedelta(days=1)), interval='1d'))["Open"]).item()    # Opening Price the morning after the split 

PreSH_mag = ((ticker_obj.history(start=(ID_date), end=(split_date), interval='1d'))["High"]).max()                              # Price of Pre-Split High
PreSH_date = ((ticker_obj.history(start=(ID_date), end=(split_date), interval='1d'))["High"]).idxmax()                          # Date of Pre-Split High
PreSH_waittime = ((PreSH_date.tz_localize(None)) - (pd.Timestamp(ID_date))).days                                                # Days after ID date that PreSH is hit

PreSL_mag = ((ticker_obj.history(start=(ID_date), end=(split_date), interval='1d'))["Low"]).min()                               # Price of Pre-Split Low
PreSL_date = ((ticker_obj.history(start=(ID_date), end=(split_date), interval='1d'))["Low"]).idxmin()                           # Date of Pre-Split Low
PreSL_waittime = ((PreSL_date.tz_localize(None)) - (pd.Timestamp(ID_date))).days                                                # Days after ID date that PreSL is hit

PoSH_mag = ((ticker_obj.history(start=(split_date), end=(end_date), interval='1d'))["High"]).max()                              # Price of Post-Split High
PoSH_date = ((ticker_obj.history(start=(split_date), end=(end_date), interval='1d'))["High"]).idxmax()                          # Date of Post-Split High
PoSH_waittime = ((PoSH_date.tz_localize(None)) - (pd.Timestamp(split_date))).days                                               # Days after split date that PoSH is hit

PoSL_mag = ((ticker_obj.history(start=(split_date), end=(end_date), interval='1d'))["Low"]).min()                               # Price of Post-Split Low
PoSL_date = ((ticker_obj.history(start=(split_date), end=(end_date), interval='1d'))["Low"]).idxmin()                           # Date of Post-Split Low
PoSL_waittime = ((PoSL_date.tz_localize(None)) - (pd.Timestamp(split_date))).days                                               # Days after split date that PoSL is hit


# Diagnostic Printing
print(PoSH_mag)
print(PoSH_date)
print(PoSH_waittime)
print(historical_data)
```
