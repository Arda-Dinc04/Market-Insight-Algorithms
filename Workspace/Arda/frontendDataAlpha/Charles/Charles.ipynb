```python
import pandas as pd

# Function to extract the 'b' component of the split ratio
def get_split_ratio_b(split_ratio):
    try:
        return int(split_ratio.split(':')[1].strip())
    except Exception as e:
        print(f"Error parsing split ratio '{split_ratio}': {e}")
        return 1

# Path to the input CSV file
file_path = 'initial_with_split_ratio.csv'  # Change this to the actual file path

# Read the CSV file
df = pd.read_csv(file_path)

# Calculate the 'b' component of the split ratio
df['Split Ratio B'] = df['Split Ratio'].apply(get_split_ratio_b)

# Remove the '$' sign from the 'Sell Price' and 'Cost Basis (CB)' columns and convert to float
df['Sell Price'] = df['Sell Price'].replace('[\$,]', '', regex=True).astype(float)
df['Cost Basis (CB)'] = df['Cost Basis (CB)'].replace('[\$,]', '', regex=True).astype(float)

# Calculate the adjusted cost basis
df['Adjusted Cost Basis'] = df['Cost Basis (CB)'] * df['Split Ratio B']

# Calculate the percentage difference
df['Difference'] = ((df['Sell Price'] - df['Adjusted Cost Basis']) / df['Adjusted Cost Basis']) * 100

# Round the 'Difference' column to 2 decimal places
df['Difference'] = df['Difference'].round(2)

# Calculate the average percentage difference
average_percentage_difference = df['Difference'].mean().round(2)

# Display the average percentage difference
print(f"\nAverage Percentage Difference: {average_percentage_difference}%")

# Drop the intermediate 'Split Ratio B' column
df.drop(columns=['Split Ratio B'], inplace=True)


# Save the updated DataFrame to a new CSV file
df.to_csv(file_path, index=False)

# Display the updated DataFrame
print(f"\nUpdated DataFrame saved to '{output_file_path}':")
print(df)
```
